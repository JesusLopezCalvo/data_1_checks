```python
import pandas as pd
```

```python
# Read in the data
df = pd.read_csv(r'C:\Users\Jesus\Desktop\jesugit\data_1_checks\Hospital_Service_Area_2021.csv', low_memory=False)
zipdf = pd.read_excel(r'C:\Users\Jesus\Desktop\jesugit\data_1_checks\uszips.xlsx')
```

```python
# Cleaning the data of "*" values
df = df[(df.TOTAL_DAYS_OF_CARE != "*") & (df.TOTAL_CHARGES != "*") &(df.TOTAL_CASES != "*") ]
# Keeping only the KY zip codes
KYdf = zipdf[zipdf.state_id == "KY"]
```

```python
# Convert the data to string to merge
KYdf['zip'] = KYdf['zip'].astype(str)

# Changing the column names to merge
df.columns = ['Medicare_CCN', 'zip', 'Total_days',
       'Total_charges', 'Total_cases']
```

```python
# Merge the data
df1 = df.merge(KYdf, on='zip')
```
